```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Ocene = [[1, "Una",    "Zorić",       5, 4, 2, 3, 3],
         [2, "Marta",  "Stojanović",  5, 4, 5, 5, 4],
         [3, "Ognjen", "Novaković",   3, 4, 3, 4, 4],
         [4, "Petra",  "Kaličanin",   5, 5, 5, 5, 5],
         [5, "Novak",  "Đorđević",    2, 2, 2, 3, 3],
         [6, "Đurđe",  "Martać",      1, 1, 1, 3, 1],
         [7, "Kalina", "Ognjenović",  5, 5, 4, 4, 2],
         [8, "Vanja",  "Čolović",     5, 3, 5, 4, 2]]
```

```python
TabOcene = pd.DataFrame(Ocene)
TabOcene.columns=["Rbr", "Ime", "Prezime", "Biologija", 
                  "Istorija", "Geografija", "Fizika", "Hemija"]
```

```python
TabOcene
```

```python
TabOcene["Prosek"]= ""
TabOcene["Prosek"] = TabOcene.loc[:,"Biologija":"Hemija"].mean(axis = 1)
```

```python
TabOcene
```

```python
plt.bar(TabOcene["Ime"], TabOcene["Hemija"], label="Hemija", color="y")
plt.plot(TabOcene["Ime"], TabOcene["Prosek"], label="Prosek", color="r")
plt.title("Poređenje ocena iz hemije i proseka učenika", size=15)
plt.legend()
plt.show()
plt.close()
```

```python

plt.figure(figsize=(10,5))
plt.bar(TabOcene["Prezime"], TabOcene["Istorija"], label="Istorija",color="m")
plt.title("ocena iz Istorije", size=15)
plt.xlabel("Prezimena", size=8)
plt.legend()
plt.show()
plt.close()
```
