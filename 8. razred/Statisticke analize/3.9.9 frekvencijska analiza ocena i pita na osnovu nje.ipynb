```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Ocene = [[1, "Una",    "Zorić",       5, 4, 2, 3, 3],
         [2, "Marta",  "Stojanović",  5, 4, 5, 5, 4],
         [3, "Ognjen", "Novaković",   3, 4, 3, 4, 4],
         [4, "Petra",  "Kaličanin",   5, 5, 5, 5, 5],
         [5, "Novak",  "Đorđević",    2, 2, 2, 3, 3],
         [6, "Đurđe",  "Martać",      1, 1, 1, 3, 1],
         [7, "Kalina", "Ognjenović",  5, 5, 4, 4, 2],
         [8, "Vanja",  "Čolović",     5, 3, 5, 4, 2]]
```

```python
TabOcene = pd.DataFrame(Ocene)
TabOcene.columns=["Rbr", "Ime", "Prezime", "Biologija", 
                  "Istorija", "Geografija", "Fizika", "Hemija"]
```

```python
TabOcene
```

```python
TabOcene["Prosek"]= ""
TabOcene["Prosek"] = TabOcene.loc[:,"Biologija":"Hemija"].mean(axis = 1)
```

```python
TabOcene
```

```python
#Frekvencijska analiza ocena iz hemije
OceneIzHemije = TabOcene["Hemija"].value_counts()
OceneIzHemije
```

```python
#Rezultat frekvencijske analize
```

```python
# OceneIzHemije.values su u piti a OceneIzHemije.index su u legendi
```

```python
plt.pie(OceneIzHemije.values)
plt.title("Odnos ocena iz hemije", size=15)
plt.legend(OceneIzHemije.index, bbox_to_anchor=(1,0.75))
plt.show()
plt.close()
```
